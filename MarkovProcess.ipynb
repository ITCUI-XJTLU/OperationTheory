```python
import numpy as np
```

```python
# upload data
tran = {0:np.array([[0.7 ,0.2 ,0.1],
                    [0.35,0.6 ,0.05],
                    [0.1 ,0.2 ,0.7]]),
        1:np.array([[0.5 ,0.3 ,0.2],
                    [0.2 ,0.6 ,0.2],
                    [0.15,0.4 ,0.45]]),
        2:np.array([[0.6 ,0.1 ,0.3],
                    [0.25,0.15 ,0.6],
                    [0.4 ,0   ,0.6]])}

reward = {0:np.array([[3 ,4 ,25],
                      [-2 ,10 ,22],
                      [-10, 8,10]]),
          1:np.array([[1 ,5 ,10],
                      [-5 ,4 ,9],
                      [2 ,8,12]]),
          2:np.array([[-12,5 ,30],
                      [-10,7 ,12],
                      [-10,13,40]])}

states = [0,1,2]
```

```python
result = np.array([[0,0,0]])
stages = np.arange(1,5)
stages = stages[::-1]

for stage in stages:
  stage_profit = []
  for state in states:
    print('stage: ',stage,'   state:',state)
    optimal={'state':0,'profit':'initial'}
    for action in tran.keys():
      probability = tran[action]
      probability = probability[state]
      profit = reward[action]
      profit = profit[state]
      past_value = result[len(result)-1]
      #print(probability,profit,past_value)

      profit = np.sum(probability*(profit+past_value))
      profit= format(profit, '.3f')
      profit= float(profit)
      print('action:',action,'    profit:',profit)
      if optimal['profit']=='initial':
        optimal['state'] = action
        optimal['profit'] = profit
        #print('1111')
      elif profit < optimal['profit']:
        optimal['state'] = action
        optimal['profit'] = profit

    print('The best choice is:', optimal['state'],'   the cost:',optimal['profit'])
    print()
    stage_profit.append(optimal['profit'])
  result = np.append(result,np.reshape(stage_profit,[1,len(stage_profit)]),axis=0)

result
```
